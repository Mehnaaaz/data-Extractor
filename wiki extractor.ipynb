```python
from selenium import webdriver
from selenium.webdriver.common.by import By
```

```python
dr=webdriver.Chrome()
```

```python
#Visiting website
dr.get("https://www.wikipedia.org/")
```

```python
wiki_data=dr.find_element(By.XPATH,"//body//main")
```

```python
print(wiki_data.text)
```

```python
wiki_search=dr.find_element(By.XPATH,"//input[@id='searchInput']")
```

```python
import time
time.sleep(1) 
```

```python
wiki_search.click()
```

```python
from selenium.webdriver.common.keys import Keys
wiki_search.send_keys("Margot Robbie")
wiki_search.send_keys(Keys.RETURN)
```
