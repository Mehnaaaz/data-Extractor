```python
from selenium import webdriver
from selenium.webdriver.common.by import By
```

```python
dr=webdriver.Chrome()
```

```python
#Visiting website
dr.get("https://www.flipkart.com/")
```

```python
#for finding element use find_element function
mobile=dr.find_element(By.XPATH,"//a[@aria-label='Mobiles']//div//div[@class='YBLJE4']")
```

```python
mobile.click()
```

```python
mobile_data=dr.find_element(By.XPATH,"//div[@class='JxFEK3 _48O0EI']")
```

```python
print(mobile_data.text)
```
